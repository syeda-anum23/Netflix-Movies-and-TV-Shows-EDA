## Basic Exploration of Netflix dataset

```python
import pandas as pd
import numpy as np
import matplotlib as mp
```

```python
df= pd.read_csv('netflix_titles2.csv')
df.head()
```

```python
df.info()
df.describe()
```

```python
#checking how many rows are null in total in each column
df.isna().sum()
```

## Data Cleaning

```python
#filling in missing values
df= df.fillna({'director': 'Unknown', 'cast': 'Unknown', 'country': 'Unknown'})
df
```

```python
#filling date with most common date with mode()
comm_date = df['date_added'].mode()[0]
df['date_added'] = df['date_added'].fillna(comm_date)
df['date_added']
```

```python
#removing duplicates in dataframe
df.drop_duplicates()
```

```python
#standardize the columns removing spaces
for col in df.select_dtypes(include='object').columns:
    df[col]=df[col].str.strip()
```

```python
#convert the date-added column to datetime format
df['date_added']= pd.to_datetime(df['date_added'], format='mixed')
df['date_added']
```

## Feature Engineering

```python
#adding two new columns extracting from date_added
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

```python
#creates new column for Movie vs Tv Shows
df['content_type'] = df['type'].apply(lambda x: 'Movie' if x=='Movie' else 'TV Show')
df.head()
```

```python
#Create a new column duration_in_minutes (for movies only) and seasons (for TV shows)

df['duration_num'] = df['duration'].str.extract(r'(\d+)').astype(float)

# For Movies → store numbers in "duration_in_minutes"
df['duration_in_minutes']= df['duration_num'].where(df['type']== 'Movie')

# For TV Shows → store numbers in "seasons"
df['seasons'] = df['duration_num'].where(df['type']== 'TV Show')
df= df.drop(columns=['duration_num'])
df.head()
```

```python
#Standardize rating column into categories (e.g., "Kids", "Teens", "Adults")
def categorize_rating(rating):
    if rating in ['G', 'TV-Y', 'TV-Y7','TV-G']: 
        return 'Kids'
    elif rating in ['PG', 'PG-13', 'TV-PG', 'TV-14']:
        return 'Teens'
    elif rating in ['R', 'NC-17', 'TV-MA']:
        return 'Adults'
    else:
        return 'Unknown'
df['rating_category']= df['rating'].apply(categorize_rating)
df.head()
```

## Exploratory Data Analysis (EDA)

```python
# How many movies vs TV shows are there?
df['type'].value_counts()
```

```python
# Which country produces the most Netflix content?
df['country'].value_counts().head() #top 5 countries
```

```python
# Which year saw the most releases added on Netflix?

df['year_added'].value_counts().head()
```

```python
# What are the top 10 genres on Netflix?

#Splits multiple genres (like "Dramas, Comedies") into separate rows and counts them.
df['listed_in'].str.split(',').explode().str.strip().value_counts().head(10)
```

```python
# Who are the top 10 most frequent directors?
df['director'].value_counts().head(10)
```

```python
# What’s the distribution of ratings (PG, TV-MA, etc.)?
df['rating'].value_counts().plot(kind='bar')
```

```python
# Which actors appear the most in Netflix titles?
df['cast'].dropna().str.split(',').explode().str.strip().value_counts().head(10)
```

```python
# Trend of Netflix content growth year by year.
df.groupby('year_added')['show_id'].count().plot(kind='line', marker= 'o')
```

## Advanced Tasks

```python
# Find countries that produce more TV shows than movies.

# Group and count Movies vs TV Shows per country
country_counts = df.groupby(['country', 'type']).size().unstack(fill_value=0)

# Filter countries where TV Shows > Movies
more_tv = country_counts[country_counts['TV Show'] > country_counts['Movie']]

print(more_tv.head(10))  # print top 10 for quick view
```

```python
# Find the longest movie and TV show with most seasons

# Movie with longest duration in mins
longest_movie = df[df['type']== 'Movie'].sort_values('duration_in_minutes', ascending=False).head(1)
print(longest_movie[['title', 'type', 'duration', 'country']].head(1),'\n')

# TV Show with most seasons
longest_show = df[df['type']=='TV Show'].sort_values('seasons', ascending=False).head(1)
print(longest_show[['title', 'type', 'duration', 'country']].head(1))
```
